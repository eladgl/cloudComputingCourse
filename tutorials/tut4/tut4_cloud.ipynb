```python
import pandas as pd

# Replace 'path_to_your_file' with the actual path in your Google Drive
file_path = '/content/drive/MyDrive/data.json'
data = pd.read_json(file_path)
```

```python
import ipywidgets as widgets
from IPython.display import display

# Create dropdowns for filtering
user_dropdown = widgets.Dropdown(
    options=data['User'].unique(),
    description='User:',
    disabled=False
)

document_dropdown = widgets.Dropdown(
    options=data['Document'].unique(),
    description='Document:',
    disabled=False
)

# Display the widgets
display(user_dropdown, document_dropdown)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def update_plot(user, document):
    filtered_data = data[(data['User'] == user) & (data['Document'] == document)]

    # Example plot: Count of actions per tab
    plt.figure(figsize=(10, 6))
    sns.countplot(data=filtered_data, x='Tab', order=filtered_data['Tab'].value_counts().index)
    plt.title(f'Actions for {user} in {document}')
    plt.xticks(rotation=45)
    plt.show()

# Connect widgets to the update function
widgets.interactive(update_plot, user=user_dropdown, document=document_dropdown)
```

```python
def plot_actions_over_time(user, document):
    filtered_data = data[(data['User'] == user) & (data['Document'] == document)]
    filtered_data['Time'] = pd.to_datetime(filtered_data['Time'])

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=filtered_data, x='Time', y='Description', marker='o')
    plt.title(f'Actions Over Time for {user} in {document}')
    plt.xticks(rotation=45)
    plt.show()

# Add additional widgets or modify existing ones for this plot
widgets.interactive(plot_actions_over_time, user=user_dropdown, document=document_dropdown)
```
